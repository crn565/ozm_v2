# Creacion de los ficheros de datos en formato csv a partir de los h5

A  continuacion vamos a hacer una serie  de manipulaciones con los datos aportados en los ficheros de datos en formato h5.

Resumidamente  estos son los pasos que vamos a seguir:


- En primer lugar cargaremos los ficheros en un dataset 
- En segundo lugar calcularemos la potencia aparente en funcion de dos columnas ( pot real y pot reactiva)
- En tercer lugar eliminaremos las columnas qeu no vamos a necesitar
- Seguidamente reordenamos las columnas en el orden que necesitamos
- Finalmente salvamos el dataset obtenido en un fichero en formato csv

## Generación del fichero de datos a partir del fichero fan.h5  

```python
import pandas as pd
import numpy as np
df_datos_agregados = pd.read_hdf ('./data/ozm1/fan.hdf5')

#df_datos_agregados [['Time','Frequency','Unbalance','Flag','Voltage','Current','Active_Power','Reactive_Power','Power_Factor','Phi','Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics','target']]
#calculo de la potencia aparente
df_datos_agregados['a2']=pow(2,df_datos_agregados['Active_Power'])
df_datos_agregados['b2']=pow(2,df_datos_agregados['Reactive_Power'])
df_datos_agregados['c']=df_datos_agregados['a2'] + df_datos_agregados['b2']
df_datos_agregados ['VA']=np.sqrt((df_datos_agregados['c']))

#renombramos columnas
#df_datos_agregados ['timestamp']=df_datos_agregados ['Time']
#df_datos_agregados ['W']=df_datos_agregados ['Active_Power']
#df_datos_agregados ['VAR']=df_datos_agregados ['Reactive_Power']
#df_datos_agregados ['f']=df_datos_agregados ['Frequency']
#df_datos_agregados ['VLN']=df_datos_agregados ['Voltage']
#df_datos_agregados ['PF']=df_datos_agregados ['Power_Factor']
#df_datos_agregados ['A']=df_datos_agregados ['Current']


df_datos_agregados.rename(columns={'Time':'timestamp','Active_Power':'W','Reactive_Power':'VAR' },inplace=True)
df_datos_agregados.rename(columns={'Frequency':'f','Voltage':'VLN','Power_Factor':'PF','Current':'A' },inplace=True)



df_datos_agregados.drop (['FirstTimestamp','OriginalTimestamp','Tag'],axis = 'columns', inplace=True)

df_datos_agregados.drop (['Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics'],axis = 'columns', inplace=True)

df_datos_agregados.drop (['Flag','Unbalance','a2','b2','c','Phi'],axis = 'columns', inplace=True)



df_datos_agregados=df_datos_agregados [['timestamp','W','VAR','VA','f','VLN','PF','A']]


#df_datos_agregados.drop (['FirstTimestamp','OriginalTimestamp','Tag','Time','Voltage','Current','Frequency','Flag','Unbalance','Power_Factor'],axis = 'columns', inplace=True)
#df_datos_agregados.drop (['Phi','Active_Power','Reactive_Power','Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics'],axis = 'columns', inplace=True)
#eliminamos calculos intermedios
#df_datos_agregados.drop (['a2','b2','c'],axis = 'columns', inplace=True)


print (df_datos_agregados)

df_datos_agregados.to_csv('fan.csv')


```
